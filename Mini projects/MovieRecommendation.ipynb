Develop a movie recommendation model using the scikit-learn library in python. 

```python
# Import libraries needed for the movie recommendation system.
# pandas and numpy help with data handling.
# CountVectorizer converts text into numeric vectors.
# cosine_similarity compares movies based on those vectors.
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load the movie dataset directly from a GitHub CSV URL.
# The dataset contains movie details like title, cast, genres, director, and keywords.
df = pd.read_csv("https://raw.githubusercontent.com/rashida048/Some-NLP-Projects/master/movie_dataset.csv")
```

```python
# Display the first five rows to understand the structure of the dataset.
df.head()
```

**Output Explanation:**

This table shows the first five movies from the dataset. It helps us understand what columns are available before building the recommendation system. Important columns include `title`, `genres`, `keywords`, `cast`, and `director`, because these are later used to find movies similar to each other.


```python
# Select the text-based columns that will be used to compare movies.
# These features describe the story, actors, category, and director.
features = ['keywords','cast','genres','director']
```

**Output Explanation:**

This output does not display a table. It stores the column names that will be used for recommendation. The model compares movies using `keywords`, `cast`, `genres`, and `director`.


```python
# Combine selected text features into one single string for each movie.
# This makes it easier to convert all movie information into one vector.
def combine_features(row):
    return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']
```

```python
# Replace missing values in selected columns with an empty string.
# This prevents errors when combining text values.
for feature in features:
    df[feature] = df[feature].fillna('')

# Create a new column that contains all selected features combined together.
df["combined_features"] = df.apply(combine_features,axis=1)
```

**Output Explanation:**

This cell creates a new column called `combined_features`. Each movie now has one combined text value made from keywords, cast, genres, and director. This combined text is what the recommendation system uses to compare movies.


```python
# Create a CountVectorizer object to convert text into numbers.
cv = CountVectorizer() 
# Fit the vectorizer on combined_features and transform the text into a count matrix.
count_matrix = cv.fit_transform(df["combined_features"])
```

**Output Explanation:**

This output is not shown as a normal table because `count_matrix` is a numeric matrix. It represents the words from `combined_features` as numbers. The machine learning algorithm needs this numeric form to calculate similarity between movies.


```python
# Calculate similarity between every pair of movies using cosine similarity.
# A higher similarity score means two movies are more alike.
cosine_sim = cosine_similarity(count_matrix)
```

**Output Explanation:**

This cell creates a similarity matrix. Each value in the matrix tells how similar one movie is to another movie. A higher value means the movies have more similar keywords, cast, genres, or director.


```python
# Helper function: get a movie title using its dataframe index.
def get_title_from_index(index):
    return df[df.index == index]["title"].values[0]

# Helper function: get a movie index using its title.
def get_index_from_title(title):
    return df[df.title == title]["index"].values[0]
```

```python
# Choose the movie for which we want recommendations.
movie_user_likes = "Avatar"
# Find the index number of the selected movie.
movie_index = get_index_from_title(movie_user_likes)
# Get similarity scores between the selected movie and all other movies.
# enumerate keeps both the movie index and its similarity score.
similar_movies = list(enumerate(cosine_sim[movie_index]))
```

**Output Explanation:**

This cell gets the similarity scores for the selected movie, `Avatar`. Each item contains a movie index and its similarity score compared with `Avatar`.


```python
# Sort movies by similarity score in descending order.
# [1:] removes the first result because it is the same movie itself.
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
```

**Output Explanation:**

This cell sorts all movies from most similar to least similar. The first movie is removed because it would be `Avatar` itself, which is always the most similar to `Avatar`.


```python
# Print the top similar movie recommendations.
i=0
print("Top 5 similar movies to "+movie_user_likes+" are:\n")
# Loop through sorted movies and print their titles.
for element in sorted_similar_movies:
    print(get_title_from_index(element[0]))
    i=i+1
    # Stop after printing 6 movies because the condition is i > 5.
    if i>5:
        break
```

**Output Explanation:**

This output shows the recommended movies for the selected movie `Avatar`. These movies are printed because they have the highest cosine similarity scores with `Avatar` based on the selected features: keywords, cast, genres, and director.

If you are explaining this result, you can say: the system recommends these movies because their content information is most similar to `Avatar`.
